```python
import os
import xml.etree.ElementTree as et
```

```python
#retorna lista con todos los .xml en la carpeta ó path llamado <path>
def filenames(path):
    return [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f)) if f[-3:]=='xml']

files=filenames('WSI')
files
```

```python
#retorna una lista donde cada elemento son las poligonales registradas en fn
def get_vert(fn):
    tree = et.parse(fn)
    root = tree.getroot()
    n,v = len(root[0][1]), []
    for i in range(1,n):
        x=root[0][1][i][1]
        v.append([[c.get('X'), c.get('Y')] for c in x])
        #v.append([[c.get('X'), c.get('Y'), c.get('Z')] for c in x]) # Z siempre era 0. Lo omitimos
    return v
```

```python
d=get_vert('WSI/I-2021-5616 A1.xml')
```

```python
len(d), len(d[0]), len(d[1]), len(d[2])#, len(d[3])
```

```python
# devuelve diccionario con Todas las poligonales para cada .xml, dado el path de la carpeta con los xml
def get_all_vert(path):
    files = [f for f in os.listdir(path) if os.path.isfile(os.path.join(path, f)) if f[-3:]=='xml']
    return {f: get_vert(os.path.join(path,f)) for f in files}
```

```python
d = get_all_vert('WSI')
```

```python
len(d['I-2021-5616 A1.xml']), len(d['I-2021-5616 A3.xml']), len(d['I-2021-5616 B6.xml'])
```
